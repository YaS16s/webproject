```python
from google.colab.patches import cv2_imshow
img = cv2.imread("/content/jose-g-ortega-castro-UGAyT0Xll0k-unsplash.jpg")
y=img.shape[0]
x=img.shape[1]
cy=y//2
cx=x//2
def draw_square(img,cy,cx,size,color):
  x1=cx-size #x1=st_x
  x2=cx+size  #x2=end_x
  y1=cy-size  #y1=st_y
  y2=cy+size  #y2=end_y
  img[y1:y2,x1:x2]=color
  return img


draw_square(img,cy,cx+100,100,[0,0,255])
draw_square(img,cy,cx+100,50,[0,255,0])

cv2_imshow(img)










```

# video work

```python
import cv2
from google.colab.patches import cv2_imshow

cap = cv2.VideoCapture("/content/From Main Klickpin CF- Skydiving With The Whole Group🤸♀️🔥😎 - 3lLIjvKoG.mp4")
fram_c = 0
while True:
  ret,frame=cap.read() #ret = return flag << bhuila jais na >>
  if ret==False:
    break
  cv2_imshow(frame)

  fram_c +=1
  if fram_c >= 5:
    break

  if cv2.waitKey(1) == 27: #27=esc key , 27 ascii number of esc
    break

cap.release()

cv2.destroyAllWindows()
```

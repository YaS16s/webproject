```python
from google.colab.patches import cv2_imshow
img = cv2.imread("/content/jose-g-ortega-castro-UGAyT0Xll0k-unsplash.jpg")
y=img.shape[0]
x=img.shape[1]
cy=y//2
cx=x//2
def draw_square(img,cy,cx,size,color):
  x1=cx-size #x1=st_x
  x2=cx+size  #x2=end_x
  y1=cy-size  #y1=st_y
  y2=cy+size  #y2=end_y
  img[y1:y2,x1:x2]=color
  return img


draw_square(img,cy,cx+100,100,[0,0,255])
draw_square(img,cy,cx+100,50,[0,255,0])

cv2_imshow(img)










```

```python
import cv2
from google.colab.patches import cv2_imshow # Added for cv2_imshow

cap=cv2.VideoCapture("/content/From Main Klickpin CF- Skydiving With The Whole Group🤸♀️🔥😎 - 3lLIjvKoG.mp4")

if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

ret,frame1=cap.read()
ret,frame2=cap.read()

while ret:
  dif=cv2.absdiff(frame1,frame2)
  gray=cv2.cvtColor(dif,cv2.COLOR_BGR2GRAY) # Corrected cvtcolor to cvtColor
  blur=cv2.GaussianBlur(gray,(5,5),0) #noise remove er jonjo blur use hoy
  _,thresh = cv2.threshold(blur,20,255,cv2.THRESH_BINARY)

  cv2_imshow(thresh)

  #update frames
  frame1=frame2
  ret,frame2=cap.read()

  if not ret:
    break

  if cv2.waitKey(1) & 0xFF == ord('q'): # Corrected watkey to waitKey
    break

cap.release() # Corrected cap_release() to cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from google.colab.patches import cv2_imshow
cap = cv2.VideoCapture("/content/From Main Klickpin CF- Skydiving With The Whole Group🤸♀️🔥😎 - 3lLIjvKoG.mp4")

ret,frame1=cap.read()
ret,frame2=cap.read()
while ret:
  diff = cv2.absdiff(frame1,frame2)
  gray = cv2.cvtColor(diff,cv2.COLOR_BGR2GRAY)
  blur= cv2.GaussianBlur(gray,(5,5),0)
  _,thresh = cv2.threshold(blur,20,255,cv2.THRESH_BINARY)
  contours,_=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

  for cnt in contours:
    x,y,w,h = cv2.boundingRect(cnt)

    cv2.rectangle(frame1,(x,y),(x+w , y+h),(0,0,255), 3 )

  cv2_imshow(frame1)
  frame1=frame2
  ret,frame2=cap.read()
  if cv2.waitKey(1) & 0xFF == ord('q'):
    break
cap.release()
cv2.destroyAllWindows()
```
